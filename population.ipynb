```python
# import population df

csv_file = "Resources/population.csv"
population_df = pd.read_csv(csv_file, encoding="ISO-8859-1")
population_df.head()
```

```python
#drop unnecessary columns from income df

income_df = income_df.drop(columns=['SpecialNotes','TableName'])
income_df
```

```python
#drop blank from income df table

income_df = income_df.dropna(how='any')
income_df

#country id??
```
